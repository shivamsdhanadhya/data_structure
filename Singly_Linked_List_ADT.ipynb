- Linked List:
Linked List is a data structure is used to store data collection. It has following properties:
* Successive elements are connected by pointers.
* The last element points to NULL.
* Can grow or shrink during the execution of program.
* Can be made as long as required.
* does not waste the memory but takes some extra space for storing pointers.

Main Linked List Operations:
* Insert 
* Delete

Auxiliary Linked List Operations:
* Delete List
* Count
* Find Nth node from the end of the list

Arrays vs Linked List:
 * Advantages of arrays:
   i]  easy to use
   ii] faster access (O(1))
 
 * Disadvantages of arrays:
   i]   fixed Size
   ii]  memory wastage
   iii] complex insertion (position based)

 * Advantages of Linked List:
   i] Dynamic allocation for insertion and deletion
 
 * Disadvantages of Linked List:
   i]  access time to individual elements
   ii] sometimes LL are hard to manipulate

```python
# Linked-List class defining Single-linked list
class LinkedList:    
    #Constructor
    def __init__(self):
        self.head = None

    # Method to insert a node at the beginning of singly linked list    
    def insert_at_begin(self, data):
        node = Node(data)
        node.set_next(self.head)
        self.head = node
        
    # Method to  insert a node at the end of the singly linked list    
    def insert_at_end(self, data):
        node = Node(data)
        current = self.head
        while current.get_next() != None:
            current = current.get_next()
        current.set_next(node) 
        
    # Method to insert a node at a given position
    def insert_at_pos(self, data, pos):
        current = self.head
        pos_counter = 1
        if pos == 1:
            self.insert_at_begin(data)
        elif pos == self.size():
            self.insert_at_end(data)
        else:
            while pos_counter < pos-1:
                current = current.get_next()
                pos_counter += 1
            node = Node(data)
            next_for_node = current.get_next()
            next_for_curr = node
            current.set_next(next_for_curr)
            node.set_next(next_for_node)

    # Method to get size of the singly linked-list    
    def size(self):
        current = self.head
        size = 0 
        while current != None:
            size += 1
            current = current.get_next()
        return size
    
    # Method to print singly linked-list from begin-end
    def print_linked_list(self):
        content = ''
        current = self.head
        while current != None:
            content = content + current.get_data() + ' --> ' 
            current = current.get_next()
        content += 'None'
        return content
            

# Node class defining a node in single linked list
class Node:
    #Constructor
    def __init__(self, data= None):
        self.data = data
        self.next_n = None
    #Method to get data from a node
    def get_data(self):
        return self.data
    #Method to to getnext item 
    def get_next(self):
        return self.next_n
    #Method to set data in a node
    def set_data(self, data):
        self.data = data
    #Method to set nest in a node    
    def set_next(self, next_n):
        self.next_n = next_n
    #Method returns true, if node has next 
    def has_next(self):
        return self.next_n != None
    

singly_linked_list = LinkedList()
singly_linked_list.head = Node('Mon')
singly_linked_list.insert_at_end('Tue')
singly_linked_list.insert_at_end('Wed')
singly_linked_list.insert_at_end('Fri')
singly_linked_list.insert_at_end('Sat')
singly_linked_list.insert_at_end('Sun')
singly_linked_list.insert_at_pos('Thu', 4)
print('Size of LinkedList is: {0}'.format(singly_linked_list.size()))
print('\nContents of LinkedList are:\n{0}\n'.format(singly_linked_list.print_linked_list()))
```
